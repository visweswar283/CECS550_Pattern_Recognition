## Part 8: Extra Points

* We want to perform the same operations but on the columns of the original image. How would you do the matrix algebra? Plot the scrambled image for the proof of concept.


```python
# Step 1: Create random column indices for shuffling
num_rows, num_cols, num_channels = image.shape

# Shuffle columns
col_indices = np.arange(num_cols)
np.random.shuffle(col_indices)

# Step 2: Create a transformation matrix for column permutations (T_cols)
T_cols = np.eye(num_cols)[:, col_indices]

# Step 3: Scramble the original image using T_cols for column permutations
scrambled_image = np.zeros_like(image)

for channel in range(num_channels):
    scrambled_image[:, :, channel] = np.dot(image[:, :, channel], T_cols.T)

# Step 4: Plot the scrambled image to demonstrate the concept
plt.imshow(scrambled_image.astype(np.uint8))
plt.title('Scrambled Image (Columns)')
plt.axis('off')
plt.show()
```

## Extra Points

* Now can you perform the scrambling on __both the columns and rows__ of the original image?  Write the code and plot the scrambled image. Then invert the operations and show that you can recover the original image. 

```python
# Step 1: Create random row and column indices for shuffling
num_rows, num_cols, num_channels = image.shape

# Shuffle rows
row_indices = np.arange(num_rows)
np.random.shuffle(row_indices)

# Shuffle columns
col_indices = np.arange(num_cols)
np.random.shuffle(col_indices)

# Step 2: Create transformation matrices for row and column permutations
T_rows = np.eye(num_rows)[row_indices]
T_cols = np.eye(num_cols)[:, col_indices]

# Step 3: Scramble the original image using both T_rows and T_cols
scrambled_image = np.zeros_like(image)

for channel in range(num_channels):
    scrambled_image[:, :, channel] = np.dot(np.dot(T_rows, image[:, :, channel]), T_cols)

# Step 4: Invert the operations to recover the original image
T_rows_inv = np.linalg.inv(T_rows)
T_cols_inv = np.linalg.inv(T_cols)
recovered_image = np.zeros_like(image)

for channel in range(num_channels):
    recovered_image[:, :, channel] = np.dot(np.dot(T_rows_inv, scrambled_image[:, :, channel]), T_cols_inv)

# Step 5: Plot the scrambled and recovered images
plt.subplot(1, 2, 1)
plt.imshow(scrambled_image.astype(np.uint8))
plt.title('Scrambled Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(recovered_image.astype(np.uint8))
plt.title('Recovered Image')
plt.axis('off')

plt.show()
```
